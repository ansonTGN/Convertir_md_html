Why Should We Convert PDF to Markdown?

Without Markdown, we would normally use a PDF reader to extract the content of our document.

However, this leads to a loss of information about the structure of the document, as well as a loss of information due to incorrect content conversions.

```python
# Instalar pypdf (CASO 01)
!pip install pypdf
```

```python
# Dato de ejemplo para pruebas
Dato = "./DATOS/AUrbina_CV.pdf"
```

```python
# CASO 1

from pypdf import PdfReader

reader = PdfReader(Dato)
page = reader.pages[0]
print(page.extract_text())
```

```python
# Caso 02 PyMuPDF4LLM
```

```python
!pip install pymupdf4llm
```

```python
import pymupdf4llm
import pathlib

md_text = pymupdf4llm.to_markdown(Dato)

# save to disk
pathlib.Path("output-pymudpdf4llm.md").write_bytes(md_text.encode())
```

```python
# Caso 03 Dockling IBM

# https://github.com/DS4SD/docling
```

```python
!pip install docling
```

```python
from docling.document_converter import DocumentConverter

# Ruta al documento que deseas convertir
ruta_documento = Dato  # Reemplaza con la ruta real de tu documento

# Crear una instancia del convertidor
convertidor = DocumentConverter()

# Convertir el documento
resultado = convertidor.convert(ruta_documento)

# Exportar a Markdown
contenido_markdown = resultado.document.export_to_markdown()
with open("out_Dockling.md", "w", encoding="utf-8") as archivo_md:
    archivo_md.write(contenido_markdown)

# Exportar a HTML
contenido_html = resultado.document.export_to_html()
with open("out_Dockling.html", "w", encoding="utf-8") as archivo_html:
    archivo_html.write(contenido_html)

print("Conversión completada: se han generado 'documento_convertido.md' y 'documento_convertido.html'.")
```

# Caso 04 marker

Datalab’s marker can convert PDFs and images to Markdown, JSON, and HTML. It uses deep learning models, so it works best with Nvidia GPU support.

It is open source and licensed under the GPL-3.0.

```python
!pip install -U marker-pdf==1.3.5
```

```python
from marker.converters.pdf import PdfConverter
from marker.models import create_model_dict
from marker.output import text_from_rendered

converter = PdfConverter(
    artifact_dict=create_model_dict(),
)
rendered = converter("/path/to/bench_pdf.pdf")

# save to disk
with open("marker-output.md", "w", encoding="utf-8") as myfile:
    myfile.write(rendered.markdown)
```

# Caso 05 MarkItDown

Microsoft’s MarkItDown is an open source tool that can convert various file types to Markdown and is licensed under the MIT.

```python
!pip install -U markitdown==0.0.1a3
```

```python
from markitdown import MarkItDown

md = MarkItDown()
result = md.convert("/path/to/bench_pdf.pdf")

# save to disk
with open("markitdown-output.md", "w", encoding="utf-8") as myfile:
    myfile.write(result.text_content)
```
